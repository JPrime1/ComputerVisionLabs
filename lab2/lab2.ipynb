```python
def color2gray(img):
    r = img[...,0]
    g = img[...,1]
    b = img[...,2]
    gray = r/3 + g/3 + b/3
    return gray
```

```python
def gkern(l=5, sig=1.):
    """
    creates gaussian kernel with side length `l` and a sigma of `sig`
    """
    ax = np.linspace(-(l - 1) / 2., (l - 1) / 2., l)
    gauss = np.exp(-0.5 * np.square(ax) / np.square(sig))
    kernel = np.outer(gauss, gauss)
    return kernel / np.sum(kernel)
```

```python
flower = cv.imread("flower1.jpg")
t1 = time.time()
identity = cv.filter2D(src=flower, ddepth=-1, kernel=gkern())
t2 = time.time()
cv.imshow("og", flower)
cv.imshow("identity", identity)
cv.waitKey()
cv.destroyAllWindows()
print("time in sec:",(t2-t1))
```
